```python
import concurrent.futures
from urllib.parse import urlparse
from bs4 import BeautifulSoup
from selenium import webdriver
import time

# Function to get all links from a webpage
def get_links(url):
    try:
        driver = webdriver.Chrome()  # Replace with the path to your Chrome driver
        driver.get(url)
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        links = []
        for a in soup.find_all('a', href=True):
            link = a['href']
            if link.startswith('/') or urlparse(link).netloc == urlparse(url).netloc:
                links.append(link)
        driver.quit()
        return links
    except Exception as e:
        print("Error getting links from {}: {}".format(url, e))
        return []

# Function to count the total number of videos on a webpage
def count_videos(url):
    try:
        driver = webdriver.Chrome()  # Replace with the path to your Chrome driver
        driver.get(url)
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        video_count = 0
        for video in soup.find_all('video'):
            video_count += 1
        driver.quit()
        return video_count
    except Exception as e:
        print("Error counting videos on {}: {}".format(url, e))
        return 0

# Main function to crawl links up to 5 levels and count total number of videos
def crawl_and_count_videos(url, depth=0):
    if depth > 5:
        return 0
    links = get_links(url)
    video_count = count_videos(url)
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        # Limit the number of concurrent threads to 5
        futures = [executor.submit(crawl_and_count_videos, link, depth+1) for link in links]
        for future in concurrent.futures.as_completed(futures):
            video_count += future.result()
    return video_count

# Replace 'x' with the main URL to start crawling from
main_url = 'https://www.mewatch.sg/'
total_videos = crawl_and_count_videos(main_url)
print("Total number of videos on the webpage:", total_videos)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import time

# Function to get all links from a webpage
def get_links(url):
    driver = webdriver.Chrome()  # Replace with the path to your Chrome driver
    driver.get(url)
    time.sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    links = []
    for a in soup.find_all('a', href=True):
        link = a['href']
        if link.startswith('/') or urlparse(link).netloc == urlparse(url).netloc:
            links.append(link)
    driver.quit()
    return links

# Function to count the total number of videos on a webpage
def count_videos(url):
    driver = webdriver.Chrome()  # Replace with the path to your Chrome driver
    driver.get(url)
    time.sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    video_count = 0
    for video in soup.find_all('video'):
        video_count += 1
    driver.quit()
    return video_count

# Main function to crawl links up to 5 levels and count total number of videos
def crawl_and_count_videos(url, depth=0):
    if depth > 5:
        return 0
    links = get_links(url)
    video_count = count_videos(url)
    for link in links:
        video_count += crawl_and_count_videos(link, depth+1)
    return video_count

# Replace 'x' with the main URL to start crawling from
main_url = 'https://www.mewatch.sg/'
total_videos = crawl_and_count_videos(main_url)
print("Total number of videos on the webpage:", total_videos)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import requests

def extract_images(url, max_images=5, min_image_size_kb=300):
    print("Initializing Chrome WebDriver...")
    chrome_options = Options()
    chrome_options.add_argument('--headless')

    # Use ChromeDriverManager to automatically download and manage the ChromeDriver executable
    service = ChromeService(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)

    try:
        print(f"Navigating to {url}...")
        # Navigate to the provided URL
        driver.get(url)

        # Wait for the page to load (you may adjust the timeout as needed)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
        print("Page loaded successfully.")

        # Extract all image links with size above 300 KB
        image_links = []
        for img_tag in driver.find_elements(By.TAG_NAME, 'img'):
            img_src = img_tag.get_attribute('src')

            # Fetch image data to check the size
            img_response = requests.get(img_src)
            img_size_kb = len(img_response.content) / 1024

            # If image size is above 300 KB, add to the list
            if img_size_kb > min_image_size_kb:
                image_links.append(img_src)
                print(f'Image URL (Size: {img_size_kb} KB): {img_src}')

                # If we have reached the desired number of images, break the loop
                if len(image_links) == max_images:
                    print(f"Reached the limit of {max_images} images. Exiting...")
                    break

    finally:
        # Close the WebDriver
        print("Closing Chrome WebDriver...")
        driver.quit()
        print("Chrome WebDriver closed.")

    return image_links

# Example usage
url = "https://www.urbanmonkey.com/"
extracted_images = extract_images(url, max_images=5, min_image_size_kb=300)

print("Extracted Image URLs:")
for i, img_url in enumerate(extracted_images, start=1):
    print(f"{i}. {img_url}")
```
